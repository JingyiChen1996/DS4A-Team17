## Analyzing ridership and membership Citi Bike data

### Goal 1: import ride and membership data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import date
plt.style.use('seaborn')
```

```python
# read in ridership and membership data
df_2019_q1 = pd.read_csv('data/data-cZMp8.csv')
df_2018_q1 = pd.read_csv('data/data-7OLX8.csv')
df_2017_q1 = pd.read_csv('data/data-9KAuN.csv')
df_2019_q2 = pd.read_csv('data/data-MM5kM.csv')
df_2018_q2 = pd.read_csv('data/data-tqNn6.csv')
df_2017_q2 = pd.read_csv('data/data-Whjsi.csv')
```

```python
# merge data across quarters
df_2019 = pd.concat([df_2019_q1, df_2019_q2])
df_2018 = pd.concat([df_2018_q1, df_2018_q2])
df_2017 = pd.concat([df_2017_q1, df_2017_q2])
```

```python
# rename columns 
df_2019.columns = ['date', 'trips', 'miles', "members","day_passes", "three_day_passes"]
df_2018.columns = ['date', 'trips', 'miles', "members","day_passes", "three_day_passes"]
df_2017.columns = ['date', 'trips', 'miles', "members","day_passes", "three_day_passes"]
```

```python
# convert date into datetime
df_2019['date'] = pd.to_datetime(df_2019['date'])
df_2018['date'] = pd.to_datetime(df_2018['date'])
df_2017['date'] = pd.to_datetime(df_2017['date'])
```

### Goal 2: visualize trips and miles biked

```python
# plot trips and miles
plt.plot_date(df_2019['date'], df_2019['trips'], linestyle='solid', label='trips')
plt.plot_date(df_2019['date'], df_2019['miles'], linestyle='solid', label='miles')
plt.xlabel('date')
plt.ylabel('trips taken and miles biked')
plt.tight_layout()
plt.legend()
plt.show()
```

```python
# plot trips and miles
plt.plot_date(df_2018['date'], df_2018['trips'], linestyle='solid', label='trips')
plt.plot_date(df_2018['date'], df_2018['miles'], linestyle='solid', label='miles')
plt.xlabel('date')
plt.ylabel('trips taken and miles biked')
plt.tight_layout()
plt.legend()
plt.show()
```

```python
# plot trips and miles
plt.plot_date(df_2017['date'], df_2017['trips'], linestyle='solid', label='trips')
plt.plot_date(df_2017['date'], df_2017['miles'], linestyle='solid', label='miles')
plt.xlabel('date')
plt.ylabel('trips taken and miles biked')
plt.tight_layout()
plt.legend()
plt.show()
```

Understandably, we see both the number of trips taken and the numbers of miles biked increase over the first two quarters of each of the three years. As one increases, so does the other

Note: would be interesting to pair with ride trip data to determine if trip duration increases as well

### Goal 4:  visualize normalized trip/miles data by users

```python
df_2019['users'] = df_2019['members'] + df_2019['day_passes'] + df_2019['three_day_passes']
df_2018['users'] = df_2018['members'] + df_2018['day_passes'] + df_2018['three_day_passes']
df_2017['users'] = df_2017['members'] + df_2017['day_passes'] + df_2017['three_day_passes']

df_2019['norm_trips'] = df_2019['trips']/df_2019['users']
df_2019['norm_miles'] = df_2019['miles']/df_2019['users']

df_2018['norm_trips'] = df_2018['trips']/df_2018['users']
df_2018['norm_miles'] = df_2018['miles']/df_2018['users']

df_2017['norm_trips'] = df_2017['trips']/df_2017['users']
df_2017['norm_miles'] = df_2017['miles']/df_2017['users']
```

```python
# plot trips and miles
plt.plot_date(df_2019['date'], df_2019['norm_trips'], linestyle='solid', label='trips')
plt.plot_date(df_2019['date'], df_2019['norm_miles'], linestyle='solid', label='miles')
plt.xlabel('date')
plt.ylabel('trips taken and miles biked')
plt.tight_layout()
plt.legend()
plt.show()
```

```python
# plot trips and miles
plt.plot_date(df_2018['date'], df_2018['norm_trips'], linestyle='solid', label='trips')
plt.plot_date(df_2018['date'], df_2018['norm_miles'], linestyle='solid', label='miles')
plt.xlabel('date')
plt.ylabel('trips taken and miles biked')
plt.tight_layout()
plt.legend()
plt.show()
```

```python
# plot trips and miles
plt.plot_date(df_2017['date'], df_2017['norm_trips'], linestyle='solid', label='trips')
plt.plot_date(df_2017['date'], df_2017['norm_miles'], linestyle='solid', label='miles')
plt.xlabel('date')
plt.ylabel('trips taken and miles biked')
plt.tight_layout()
plt.legend()
plt.show()
```

The number of trips and miles are increasing per user over the first two quarters. Likely as the weather gets nicer, the same number of subscribers (though increasing) want to take more and longer bike trips with Citi Bike.

Is it based on weather? weekends?

### Goal 5: checkout if data depends on the day of the week

```python
df_2019['month'] = pd.to_datetime(df_2019['date']).dt.month
df_2018['month'] = pd.to_datetime(df_2018['date']).dt.month
df_2017['month'] = pd.to_datetime(df_2017['date']).dt.month

df_2019['dayname'] = df_2019['date'].dt.day_name();
df_2018['dayname'] = df_2018['date'].dt.day_name();
df_2017['dayname'] = df_2017['date'].dt.day_name();
```

```python
df_2019_pivot = pd.pivot_table(df_2019[['month', 'dayname', 'miles']], 
                index=['month', 'dayname'], 
                aggfunc='sum')
df_2018_pivot = pd.pivot_table(df_2018[['month', 'dayname', 'trips']], 
                index=['month', 'dayname'], 
                aggfunc='sum')
df_2017_pivot = pd.pivot_table(df_2017[['month', 'dayname', 'trips']], 
                index=['month', 'dayname'],
                aggfunc='sum')
```

```python
trips_month_day_2019 = df_2019_pivot.unstack(level=0)
trips_month_day_2018 = df_2019_pivot.unstack(level=0)
trips_month_day_2017 = df_2019_pivot.unstack(level=0)
```

```python
trips_month_day_2019 = trips_month_day_2019.reindex(["Monday", "Tuesday", "Wednesday", 
                              "Thursday", "Friday", "Saturday", "Sunday"])
```

```python
f, ax = plt.subplots(figsize=(11, 15))
ax = sns.heatmap(trips_month_day_2019, annot=True, fmt="d", linewidths=.5, ax=ax)
```

Cool! As it gets warmer from January to June (months 1 to 6), more rides are taken on the weekends and fewer are taken throughout the week. Perhaps this suggests a difference in usage throughout the year; rides for work when its cold and rides for pleasure when its warm.
