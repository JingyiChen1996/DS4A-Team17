## Analyzing ridership and membership Citi Bike data

### Goal 1: import ride and membership data

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
# read in ridership and membership data
df_2019_q1 = pd.read_csv('data/data-cZMp8.csv')
df_2018_q1 = pd.read_csv('data/data-7OLX8.csv')
df_2017_q1 = pd.read_csv('data/data-9KAuN.csv')
df_2019_q2 = pd.read_csv('data/data-MM5kM.csv')
df_2018_q2 = pd.read_csv('data/data-tqNn6.csv')
df_2017_q2 = pd.read_csv('data/data-Whjsi.csv')
```

```python
# merge data across quarters
df_2019 = pd.concat([df_2019_q1, df_2019_q2])
df_2018 = pd.concat([df_2018_q1, df_2018_q2])
df_2017 = pd.concat([df_2017_q1, df_2017_q2])
```

```python
# rename columns 
df_2019.columns = ['date', 'trips', 'miles', "members","day_passes", "three_day_passes"]
df_2018.columns = ['date', 'trips', 'miles', "members","day_passes", "three_day_passes"]
df_2017.columns = ['date', 'trips', 'miles', "members","day_passes", "three_day_passes"]
```

```python
# convert date into datetime
df_2019['date'] = pd.to_datetime(df_2019['date'])
df_2018['date'] = pd.to_datetime(df_2018['date'])
df_2017['date'] = pd.to_datetime(df_2017['date'])
```

```python
df_2019.head()
```

### Goal 2: visualize trips and miles biked

```python
# plot trips and miles
plt.plot_date(df_2019['date'], df_2019['trips'], linestyle='solid', label='trips')
plt.plot_date(df_2019['date'], df_2019['miles'], linestyle='solid', label='miles')
plt.xlabel('date')
plt.ylabel('trips taken and miles biked')
plt.tight_layout()
plt.legend()
plt.show()
```

```python
# plot trips and miles
plt.plot_date(df_2018['date'], df_2018['trips'], linestyle='solid', label='trips')
plt.plot_date(df_2018['date'], df_2018['miles'], linestyle='solid', label='miles')
plt.xlabel('date')
plt.ylabel('trips taken and miles biked')
plt.tight_layout()
plt.legend()
plt.show()
```

```python
# plot trips and miles
plt.plot_date(df_2017['date'], df_2017['trips'], linestyle='solid', label='trips')
plt.plot_date(df_2017['date'], df_2017['miles'], linestyle='solid', label='miles')
plt.xlabel('date')
plt.ylabel('trips taken and miles biked')
plt.tight_layout()
plt.legend()
plt.show()
```

Understandably, we see both the number of trips taken and the numbers of miles biked increase over the first two quarters of each of the three years. As one increases, so does the other

Note: would be interesting to pair with ride trip data to determine if trip duration increases as well
